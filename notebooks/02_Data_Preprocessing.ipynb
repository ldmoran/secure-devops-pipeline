```python
import pandas as pd

# Cargar dataset
df = pd.read_csv("../data/raw/vulnerability_fix_dataset.csv")

print(df.shape)
df.head()
```

Verificar duplicados:

```python
duplicados = df.duplicated().sum()

print(f"Duplicados encontrados: {duplicados}")
```

# Tercera celda

Eliminar duplicados:

```python
df = df.drop_duplicates()

print(df.shape)
```

# Cuarta celda

Eliminar registros donde fixed_code sea explicación en inglés.

```python
patrones_texto = [
    "To fix",
    "Here's",
    "Here is",
    "You should",
    "One way",
    "To prevent"
]

mask = df["fixed_code"].astype(str).str.startswith(tuple(patrones_texto))

print("Registros a eliminar:", mask.sum())

df = df[~mask]
```

# Quinta celda

Crear dataset binario.

```python
vulnerables = pd.DataFrame({
    "code": df["vulnerable_code"],
    "label": 1
})

seguros = pd.DataFrame({
    "code": df["fixed_code"],
    "label": 0
})
```

# Sexta celda

Unir ambos datasets:

```python
security_dataset = pd.concat(
    [vulnerables, seguros],
    ignore_index=True
)

security_dataset.head()
```

# Séptima celda

Mezclar aleatoriamente:

```python
security_dataset = security_dataset.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)
```

# Octava celda

Verificar balance:

```python
security_dataset["label"].value_counts()
```

# Novena celda

Guardar resultado:

```python
security_dataset.to_csv(
    "../data/processed/security_dataset.csv",
    index=False
)

print("Dataset guardado correctamente")
```
